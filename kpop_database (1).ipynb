```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("mykpopidolsdatabase.db")
cursor = conn.cursor()
```

```python
cursor.execute("DROP TABLE IF EXISTS kpopagency")
cursor.execute("DROP TABLE IF EXISTS kpopidolgroup")
cursor.execute("DROP TABLE IF EXISTS kpopgroupmembers")
conn.commit()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS kpopagency(id
    INTEGER PRIMARY KEY AUTOINCREMENT,
    agency TEXT,
    location TEXT,
    date_established DATE,
    number_of_kpop_groups_debuted INT,
    ownername TEXT
)
""")
conn.commit()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS kpopidolgroup(id
    INTEGER PRIMARY KEY AUTOINCREMENT,
    kpop_group_name TEXT,
    type TEXT,
    debut_date DATE,
    number_of_members INT,
    agencyID INT
)
""")
conn.commit()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS kpopgroupmembers(idol_member_id
    INTEGER PRIMARY KEY AUTOINCREMENT,
    stage_name TEXT,
    real_name TEXT,
    gender TEXT,
    birthday DATE,
    position TEXT,
    zodiac TEXT,
    personality_type TEXT,
    kpopgroup_id INTEGER,
    agency_id INTEGER
)
""")
conn.commit()
```
